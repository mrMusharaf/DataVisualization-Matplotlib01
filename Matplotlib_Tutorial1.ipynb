```python
#Getting Started with Matplotlib

from matplotlib import pyplot as plt
```

```python
x = [25,26,27,28,29,30]
y = [33100, 33895, 34500,35000, 36300, 36400]
plt.plot(x,y)
py_y  = [33000,33200,33400,33500,34000,35000]
plt.plot(x,py_y)

plt.show()
```

```python
#Title and x_y_Labels
plt.title('ML Engineers and Data Scientists Median Salary (PKR) by Age')
plt.xlabel("Age")
plt.ylabel("Median Salary (PKR)")
plt.plot(x,y)
plt.plot(x,py_y)

plt.show()
```

```python
#Legend()

plt.title('ML Engineers and Data Scientists Median Salary (PKR) by Age')
plt.xlabel("Age")
plt.ylabel("Median Salary (PKR)")
plt.plot(x,y)
plt.plot(x,py_y)
plt.legend(['ML Engineers','Data Scientists'])
plt.show()
```

```python
#Labels
plt.title('ML Engineers and Data Scientists Median Salary (PKR) by Age')
plt.xlabel("Age")
plt.ylabel("Median Salary (PKR)")
plt.plot(x,y, label = 'ML Engineers')
plt.plot(x,py_y, label = "Data Scientists")
plt.legend()
plt.show()
```

```python
#Format Strings
#A format string consists of a part for color, marker and line
#fmstr = '[color][line][marker]'
plt.legend('ML Engineers','Data Scientists')
plt.title('ML Engineers and Data Scientists Median Salary (PKR) by Age')
plt.xlabel("Age")
plt.ylabel("Median Salary (PKR)")
plt.plot(x,y,'b--.', label = 'ML Enigeers')
plt.plot(x,py_y,'g-o', label = "Data Scientist")
plt.legend()
plt.show()
```

```python
#Another easy to read way
plt.title('ML Engineers and Data Scientists Median Salary (PKR) by Age')
plt.xlabel("Age")
plt.ylabel("Median Salary (PKR)")
plt.plot(x,y,color = 'k', linestyle = '--', marker = 'o', label = 'ML Enigeers')
plt.plot(x,py_y,color = 'g', linestyle = '-', marker = 'D',  label = "Data Scientists")
plt.legend()
plt.show()
```

```python
#new data for Computer Vision Expert
cv_y = [33000,33500,33700,34000,34300,35000]
plt.title('ML Engineers,Data Scientists and CV Experts Median Salary (PKR) by Age')
plt.xlabel("Age")
plt.ylabel("Median Salary (PKR)")
plt.plot(x,y,color = 'b', linestyle = '--',linewidth = 3, marker = 'o', label = 'ML Enigeers') #Added width parameter
plt.plot(x,py_y,color = 'g', linestyle = '-',linewidth = 2, marker = '|',  label = "Data Scientists") #Added width parameter
plt.plot(x,cv_y,color = 'r', linestyle = '-', marker = 'D',  label = "CV Experts")
plt.legend()
plt.show()
```

```python
#Padding
plt.title('ML Engineers,Data Scientists and CV Experts Median Salary (PKR) by Age')
plt.xlabel("Age")
plt.ylabel("Median Salary (PKR)")
plt.plot(x,y,color = 'b', linestyle = '--',linewidth = 3, marker = 'o', label = 'ML Enigeers') #Added width parameter
plt.plot(x,py_y,color = 'g', linestyle = '-',linewidth = 2, marker = '|',  label = "Data Scientists") #Added width parameter
plt.plot(x,cv_y,color = 'r', linestyle = '-', marker = 'D',  label = "CV Experts")
plt.legend()
plt.tight_layout() #for better padding when needed
plt.show()
```

```python
#Grid Function to see when a certain salary limit crosses
plt.title('ML Engineers,Data Scientists and CV Experts Median Salary (PKR) by Age')
plt.xlabel("Age")
plt.ylabel("Median Salary (PKR)")
plt.plot(x,y,color = 'b', linestyle = '--',linewidth = 3, marker = 'o', label = 'ML Enigeers') #Added width parameter
plt.plot(x,py_y,color = 'g', linestyle = '-',linewidth = 2, marker = '|',  label = "Data Scientists") #Added width parameter
plt.plot(x,cv_y,color = 'r', linestyle = '-', marker = 'D',  label = "CV Experts")
plt.legend()
plt.grid(True) #Grid function added
plt.tight_layout() #for better padding when needed
plt.show()
```
